**Exercício 1:** Criando um DataFrame
Crie um DataFrame a partir de um dicionário com os seguintes dados:

Coluna "Nome": ['Ana', 'Carlos', 'Bianca', 'João']
Coluna "Idade": [25, 34, 22, 41]
Coluna "Cidade": ['São Paulo', 'Rio de Janeiro', 'Belo Horizonte', 'Curitiba']
Exiba o DataFrame completo.

```python
import pandas as pd;
```

```python
lista = {'Nome': ['Ana','Carlos','Bianca','João'],
         'Idade': [25,34,22,41],
         'Cidade': ['São Paulo','Rio de Janeiro','Belo Horizonte','Curitiba']}

tabela_exercicios = pd.DataFrame(lista);
```

```python
display(tabela_exercicios);
```

**Exercício 2:** Acessando Colunas
Com o DataFrame criado no exercício anterior, acesse a coluna "Idade" e exiba apenas os valores dessa coluna.

Acesse as colunas "Nome" e "Cidade" e exiba esses dois dados.

```python
display(tabela_exercicios[['Idade', 'Cidade']])
```

**Exercício 3:**Filtrando Dados
Com o DataFrame criado, filtre e mostre os dados de pessoas que têm idade maior que 30.

Agora, filtre e mostre os dados de pessoas que moram em São Paulo.

```python
idade = tabela_exercicios.loc[tabela_exercicios['Idade' ] >= 30]
display(idade)
```

**Exercício 4:** Adicionando Novas Colunas

**Exercício 5:** Alterando Valores
Altere o valor de idade de Carlos para 36 anos.

Altere o salário de João para 4500.



```python
tabela_exercicios.loc[tabela_exercicios['Nome'] == 'Carlos', 'Idade'] = 36
tabela_exercicios.loc[tabela_exercicios['Nome'] == 'João', 'Salario'] = 4500
display(tabela_exercicios)
```

**Exercício 6:** Removendo Colunas
Remova a coluna "Cidade" do DataFrame.

Exiba o DataFrame para verificar se a coluna foi removida.



```python
tabela_exercicios.drop('Cidade', axis=1, inplace=True)
display(tabela_exercicios)
```

**Exercício 7:** Ordenando Dados
Ordene o DataFrame com base na "Idade" em ordem crescente.

Ordene o DataFrame com base na "Salário" em ordem decrescente.

```python
tabela_exercicios.sort_values('Idade', ascending=True, inplace=True)
display(tabela_exercicios)

tabela_exercicios.sort_values('Salario', ascending=False, inplace=True)
display(tabela_exercicios)
```

**Exercício 8:** Agrupando Dados
Crie um DataFrame com as colunas "Produto" e "Preço":

Produto: ['Café', 'Leite', 'Arroz', 'Feijão', 'Café', 'Leite']
Preço: [10, 5, 8, 6, 11, 6]
Agrupe os dados por Produto e calcule o preço total (soma) de cada produto.

```python
produtos = {
    'Produto': ['Sopa','Arroz','Feijão','Café','Leite','Pera', 'Leite'],
    'Preço': [10,5,8,6,11,6,22]
}

dbprodutos = pd.DataFrame(produtos);
display(dbprodutos)
soma = dbprodutos.groupby('Produto')['Preço'].sum()
display(soma)
```
